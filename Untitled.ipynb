```python
import pandas as pd
import numpy as np

df = pd.read_csv("Social_Network_Ads.csv")
df.head()
```

```python
df.isnull().any()
```

```python
gender_df = pd.get_dummies(df['Gender'], drop_first=True)
gender_df
```

```python
gender_df = pd.get_dummies(df['Gender'], drop_first=True)
gender_df.columns = ['Male']
gender_df['Male'] = gender_df['Male'].astype(int)   # ðŸ‘ˆ this line is the fix
gender_df
```

```python
df.drop('User ID' ,axis=1, inplace=True)
```

```python
df
```

```python
df.drop('Gender' ,axis=1, inplace=True)
```

```python
df
```

```python
df=pd.concat([df,gender_df],axis=1)
```

```python
df
```

```python
X = df.iloc[:, [0, 1, 3]].values
X
```

```python
y = df.iloc[:, -2].values
y
```

```python
#train-test split 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
X_test[0]
```

```python

#feature scaling 
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_test[0]
```

```python
# Training the Naive Bayes model on the training set
from sklearn.naive_bayes import GaussianNB

classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
#predicting the test results
y_pred = classifier.predict(X_test)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score

# Accuracy score
ac = accuracy_score(y_test, y_pred)
ac

# Confusion matrix (optional but recommended)
#cm = confusion_matrix(Y_test, y_pred)
#cm
```
